## Cambiar Variables Categoricas a Variables Numericas

## Agrupando de Datos

```python
bins=[0,5,12,18,35,60,100]
```

```python
print(df["Edad"].groupby(pd.cut(df["Edad"], bins, right=False)).count())
```

<a href="https://pandas.pydata.org/pandas-docs/version/0.15/visualization.html" > Pandas Plotting Official Site </a>

```python
# KDE = Kernel Density Estimation
#df["Edad"].plot(legend=True, kind='kde')
```

```python
# Nombre del rando de Datos
names=["1","2","3","4","5","6"]
```
